```python
def createEmptyList(n):
    adj_List={}
    for i in range(n):
        adj_List[i]=[]
    return adj_List

def printList(adj_List):
    for k,v in adj_List.items():
        print(str(k)+": "+str(v))

def findInDegree(adj_list,v):
    count=0
    for u in adj_list:
        for x,w in adj_list[u]:
            if v==x:
                count=count+1
    return count

def findOutDegree(adj_list,v):
    length=len(adj_list[v])
    return length

f=open("Directed.txt","r")
vertex_edge=f.readline()
v_e=vertex_edge.split()
vertex=int(v_e[0])
edge=int(v_e[1])


adj_list=createEmptyList(vertex)
printList(adj_list)

for i in range(edge):
    u_v=f.readline()
    u_v_slpit=u_v.split()
    u=int(u_v_slpit[0])
    v=int(u_v_slpit[1])
    w=int(u_v_slpit[2])
    adj_list[u].append((v,w))
    #AdjacencyList[v].append(u) # if undirected
print()
printList(adj_list)
print(findInDegree(adj_list,4))
print(findOutDegree(adj_list,4))
```
