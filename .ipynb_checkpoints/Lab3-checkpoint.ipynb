```python
def createMatrix(n):
    m=[]
    for j in range(n):
        r=[]
        for i in range(n):
            if(i==j):
                r.append(0)
            else:
                r.append(-1)
        m.append(r)

    return m

def printMatrix(M):
    for i in M:
        print(i)
    print("\n")
    
f=open("1.txt","r")
vertex_edges=f.readline()
vertex_edges1=vertex_edges.split()
vertex=int(vertex_edges1[0])
edges=int(vertex_edges1[1])

Matrix=createMatrix(vertex)
printMatrix(Matrix)

for i in range(edges):
    e = f.readline()
    e1=e.split()
    u=int(e1[0])
    v=int(e1[1])
    w=int(e1[2])
    Matrix[u][v]=w
    for i in range(vertex):
     Matrix[i][i]=0

def findOutDeg(M,v):
    count=0
    for i in range(len(M[0])):
        if  M[v][i]!=-1 and M[v][i]!=0:
            count=count+1    
    return count

def findInDeg(M,v):
    count=0
    for i in range(len(M[0])):
        if  M[i][v]!=-1 and M[i][v]!=0:
            count=count+1    
    return count        

def MaxEdge(M):
    max_val=-1
    u=v=-1
    for i in range(len(M)):
        for j in range(len(M[0])):
            if M[i][j]!=-1 and M[i][j]>max_val:
                max_val=M[i][j]
                u=i
                v=j
    print(u,v ,":",max_val)
   
for row in Matrix:
    print(row)

print(findOutDeg(Matrix,2))
print(findInDeg(Matrix,2))
MaxEdge(Matrix)
```
