```python
def createMatrix(n):
    m=[]
    for j in range(n):
        r=[]
        for i in range(n):
            r.append(0)
        m.append(r)
    return m

def printMaretix(M):
    for i in M:
        print(i)


f=open("g.txt","r")
vertex_edges=f.readline()
vertex_edges1=vertex_edges.split()
vertex=int(vertex_edges1[0])
edges=int(vertex_edges1[1])


G=createMatrix(vertex)


for i in range(edges):
    e = f.readline()
    e1=e.split()
    u=int(e1[0])
    v=int(e1[1])
    G[u][v]=1
    #Matrix[v][u]=1

printMaretix(G)
print()
import queue

def dfs(G,s,vertex):
    stack=queue.LifoQueue() # frontier
    visited=[0]*vertex # explored set
    stack.put(s)
    visited[s]=1
    #print(visited)#
    
    while not stack.empty():
        v=stack.get()
        print(v,end=" ")
        for i in range(vertex):
            if (G[v][i]==1 and visited[i]==0):
                stack.put(i)
                visited[i]=1
                

dfs(G,0,vertex)
```

```python
def createMatrix(n):
    m=[]
    for j in range(n):
        r=[]
        for i in range(n):
            r.append(0)
        m.append(r)
    return m

def printMaretix(M):
    for i in M:
        print(i)


f=open("g.txt","r")
vertex_edges=f.readline()
vertex_edges1=vertex_edges.split()
vertex=int(vertex_edges1[0])
edges=int(vertex_edges1[1])


G=createMatrix(vertex)


for i in range(edges):
    e = f.readline()
    e1=e.split()
    u=int(e1[0])
    v=int(e1[1])
    G[u][v]=1
    #Matrix[v][u]=1

printMaretix(G)
print()
def dfs(G,s):
    stack=[] # frontier as list
    explore_set=set() # explored set
    stack.append(s)
    explore_set.add(s)
    while len(stack)!=0:
        v=stack.pop()
        print(v,end=" ")
        for i in range(len(G[0])):
            if (G[v][i]==1 and i not in explore_set):
                stack.append(i)
                explore_set.add(i)

 
dfs(G,0)
```
