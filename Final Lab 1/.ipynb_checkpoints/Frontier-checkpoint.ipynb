# Stack

```python
import queue

# push() = put()
# pop() = get() - return and remove

stack= queue.LifoQueue()

stack.put(10)
stack.put(20)
stack.put(30)
stack.put(40)
print(stack.get())
```

```python
import queue
stack= queue.LifoQueue()

stack.put(10)
stack.put(20)
stack.put(30)
stack.put(40)
print(stack.empty())
print(stack.qsize())

while not stack.empty():
    print(stack.get())

print(stack.empty())
print(stack.qsize())
```

```python
import queue
stack= queue.LifoQueue()

stack.put("Richard")
stack.put("Philip")

while not stack.empty():
    print(stack.get())
```

# Queue

```python
import queue

q=queue.Queue()
q.put(100)
q.put(10)
q.put(20)
print(q.get())
```

```python
import queue

q=queue.Queue()
q.put(100)
q.put(10)
q.put(20)

while not q.empty():
    print(q.get())
```

# Priority Queue

```python
import queue

pq=queue.PriorityQueue()

pq.put(100)
pq.put(1)
pq.put(10)

print(pq.get())
```

```python
import queue

pq=queue.PriorityQueue()

pq.put("B")
pq.put("C")
pq.put("A")
pq.put("D")

while not pq.empty():
    print(pq.get())
```
