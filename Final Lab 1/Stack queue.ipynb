```python
import queue
def dfs(G,s,vertex):
    stack=queue.LifoQueue() #frontier
    visited=[0]*vertex #exploredset
    print(visited)
    stack.put(s)
    visited[s]=1
    while not stack.empty():
        v=stack.get()
        print(G[v])
        print(v,end=" ")
        for i in range(vertex):
            if (G[v][i]==1 and visited[i]==0):
                stack.put(i)
                visited[i]=1
    
dfs(G,0,vertex)
```
