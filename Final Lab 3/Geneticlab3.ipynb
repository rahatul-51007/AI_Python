# string to attack

```python
import random
def createBoard(n):
    board=[]
    for i in range(n):
        row=[]
        for j in range(n):
            row.append(0)
        board.append(row)
    #add queen in each column
    for j in range(n):
        i=random.randint(0,n-1)# row select randomly
        board[i][j]=1
    return board
def boardtoString(b):
    n=len(b[0])
    ch=""
    for c in range(n):
        for r in range(n):
            if b[r][c]==1:
                ch+=str(r)
    return ch

def totalAttack(ch):
    n=len(ch)
    attack=0
    for i in range(n):
        for j in range(i+1,n):
            if ch[i]==ch[j]:
                attack+=1
            if abs(int(ch[i])-int(ch[j]))==abs(i-j):
                attack+=1
    return attack
    
def printBoard(b):
    for r in b:
        print(r)

Board=createBoard(8)
printBoard(Board)
ch=boardtoString(Board)
print(ch)
total=totalAttack(ch)
print(total)
```
