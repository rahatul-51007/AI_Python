```python
import random
def createBoard(n):
    board=[]
    for i in range(n):
        row=[]
        for j in range(n):
            row.append(0)
        board.append(row)
    #add queen in each column
    for j in range(n):
        i=random.randint(0,n-1)# row select randomly
        board[i][j]=1
    return board
def boardtoString(b):
    n=len(b[0])
    ch=""
    for c in range(n):
        for r in range(n):
            if b[r][c]==1:
                ch+=str(r)
    return ch
 
def find_attacking_pair(ch):
    n=len(ch)
    attack=0
    for i in range(n):
        for j in range(i+1,n):
            if ch[i]==ch[j]:
                attack+=1
            if abs(int(ch[i])-int(ch[j]))==abs(i-j):
                attack+=1
    return attack
def non_attacking_pair(ch):
    n=len(ch)
    return (n*(n-1)//2)-find_attacking_pair(ch)
 
def createPopulation(n,p):
    population=[]
    for i in range(p):
        b=createBoard(n)
        # printBoard(b)
        s=boardtoString(b)
        # print(s)
        # print()
        population.append(s)
    return population
def printBoard(b):
    for r in b:
        print(r)
def calculateFitness(population):
    fitness = {}
    for ch in population:
        fitness[ch] = non_attacking_pair(ch)
    return fitness
def find_parents(population_fitness):
    sorted_items = sorted(population_fitness.items(), key=lambda item: item[1])
    x=len(sorted_items)
    parent1 = sorted_items[x-1][0]
    parent2 = sorted_items[x-2][0]
    return parent1, parent2

def crossOver(p1,p2):
    n=len(p1)
    crossOver_point=random.randint(0,n-1)
    #crossOver_point=4
    child1=p1[:crossOver_point]+p2[crossOver_point:]
    child2=p2[:crossOver_point]+p1[crossOver_point:]
    print(crossOver_point)
    return child1,child2

def mutation(child):
    n=len(child)
    #child_list=child.split()
    #print(child_list) 
    index=random.randint(0,n-1)
    value=random.randint(0,n-1)
    child=child[:index]+value+child[index+1:]
    return child
    


   
n = 8
population = 10
pop = createPopulation(n, population)
print("Population:", pop)
population_fitness = calculateFitness(pop)
print("Fitness:", population_fitness)
p1, p2 = find_parents(population_fitness)
print(p1,p2)
c1,c2=crossOver(p1,p2)
print(c1,c2)
m=mutation(c1)
print(m)
```
