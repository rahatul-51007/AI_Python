```python
def createList(n):
    adj={}
    for i in range(n):
        adj[i]=[]
    return adj

def printList(adj):
    for i in range(len(adj)):
        print(i, end="->")
        print(adj[i])


f=open("g.txt","r")
v_e=f.readline().split()
n=int(v_e[0])
m=int(v_e[1])

G=createList(n)
print(G)

for i in range(m):
    l=f.readline().split()
    u=int(l[0])
    v=int(l[1])
    G[u].append(v)
    #adj_List[v].append(u)
print(G)
printList(G)

import queue
def dfs(G,s):
    stack=queue.LifoQueue()
    explored_set=set()
    stack.put(s)
    explored_set.add(s)
    while not stack.empty():
        next=stack.get()
        print(next,end=" ")
        for i in range(len(G[next])):
            if G[next][i] not in explored_set:
                stack.put(G[next][i])
                explored_set.add(G[next][i])

dfs(G,0)
```
