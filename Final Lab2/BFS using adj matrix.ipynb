```python
import queue

def createMatrix(n):
    m = []
    for j in range(n):
        r = []
        for i in range(n):
            r.append(0)
        m.append(r)
    return m

def printMatrix(M):
    for i in M:
        print(i)


f = open("g.txt", "r")
vertex_edges = f.readline()
vertex_edges1 = vertex_edges.split()
vertex = int(vertex_edges1[0])
edges = int(vertex_edges1[1])

G = createMatrix(vertex)

for i in range(edges):
    e = f.readline()
    e1 = e.split()
    u = int(e1[0])
    v = int(e1[1])
    G[u][v] = 1
    # G[v][u] = 1  

printMatrix(G)
print()


def Bfs(G, s, vertex):
    q = queue.Queue()  
    visited = [0] * vertex
    explored_set=set()
    q.put(s)
    explored_set.add(s)
    #visited[s] = 1

    while not q.empty():
        v = q.get()
        print(v, end=" ")
        for i in range(vertex):
            if (G[v][i] == 1 and i not in explored_set):
                q.put(i)
                explored_set.add(i)


Bfs(G, 0, vertex)
```
