```python
import random
def createBoard(n):
    board=[]
    for i in range(n):
        row=[]
        for j in range(n):
            row.append(0)
        board.append(row)
    # add queen in each column
    for j in range(n):
        i=random.randint(0,n-1) # row select randomly
        board[i][j]=1
    return board

def createChromosome(n):
    chrom=""
    for i in range(n):
        chrom+=str(random.randint(0,n-1))
    return chrom
    
def printBoard(b):
    for r in b:
        print(r)

ch=createChromosome(8)
print(ch)
Board=createBoard(8)
printBoard(Board)
```
